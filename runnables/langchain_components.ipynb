```python
import random

class DuplicateLLM:

    def __init__(self):
        print("LLM Created")

    def predict(self,prompt):
        responses = [
            "CR7 is GOAT",
            "I am the best AI Engineer",
            "I have a very inquisitive mind who wants to learn everything"
        ]

        response = random.choice(responses)

        return {"Response": response}
```

```python
llm = DuplicateLLM()
```

```python
llm.predict("Hemllo")
```

```python
class DuplicatePrompt:

    def __init__(self,input_variable, template):
        self.input_variable = input_variable
        self.template = template

    def format(self, input_dict):
        return self.template.format(**input_dict)
```

```python
prompt = DuplicatePrompt('topic',"What is the {topic}?")
```

```python
prompt = prompt.format({"topic":"a"})
```

```python
llm = DuplicateLLM()
```

```python
llm.predict(prompt)
```

```python
class DuplicateChain:

    def __init__(self,llm,prompt):
        self.llm = llm
        self.prompt = prompt

    def run(self,input_dict):
        final_prompt = self.prompt.format(input_dict)
        result = self.llm.predict(final_prompt)
        return result['Response']
```

```python
llm = DuplicateLLM()
```

```python
prompt = DuplicatePrompt('player',"What's the name of the {player}")
```

```python
chain = DuplicateChain(llm=llm,prompt=prompt)
```

```python
chain.run({"player":"CR7"})
```
