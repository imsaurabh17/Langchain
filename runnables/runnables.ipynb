```python
from abc import ABC, abstractclassmethod
import random
```

```python
class NakliRunnable(ABC):
    @abstractclassmethod
    def invoke(self):
        pass
```

```python
class NakliLLM(NakliRunnable):

    def __init__(self):
        print("LLM Created")

    def predict(self,prompt):
        response_list = [
            "AB devilliers is the Best",
            "The magic you are looking for is in the work you are avoiding",
            "Be like a water"
        ]

        response = random.choice(response_list)

        return response
    
    def invoke(self,promt):
        response_list = [
            "AB devilliers is the Best",
            "The magic you are looking for is in the work you are avoiding",
            "Be like a water"
        ]

        response = random.choice(response_list)

        return response
```
